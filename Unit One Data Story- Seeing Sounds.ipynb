## ***Stage Fright: Potential Pitfalls of Musical Visualizations***

### Although absorbable and visually appealing, one must consider the potential harm that comes with data visualization. For one, it could be argued that by converting the data into graphs, it can invalidate necessary context. Visual mediums are helpful and engaging, but it can be hard to include all the nuances that tables and datasets manage to include. For issues like causation and correlation, graphs can often misrepresent the real cause of certain connections if not clarified. One could also fall too deep into the design part of making visualizations that they lose focus on the priority: making sure data is visualized in a way all can absorb. They may make things harder to understand without realizing, too focused on enjoying the way it looks to theme despite it's inaccessibility to others. ### ***Track Seven: Musical Mentality***
### This histogram paints a picture of music's frequency on self-reported mental health impact: a composition that intertwines music and mental well-being.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('mxmh_survey_results.csv')

categories = ['Anxiety', 'Depression', 'Insomnia', 'OCD']
frequencies = [
    df['Anxiety'].value_counts().sort_index().tolist(),
    df['Depression'].value_counts().sort_index().tolist(),
    df['Insomnia'].value_counts().sort_index().tolist(),
    df['OCD'].value_counts().sort_index().tolist()
]

num_categories = len(categories)
bar_width = 0.2

positions = np.arange(len(set(df['Anxiety'].tolist())))

colors = ['#FF3366', '#FF6666', '#FF9966', '#FFCC66']

plt.figure(figsize=(10, 6))

for i in range(len(frequencies)):
    plt.bar(positions + i * bar_width, frequencies[i][:len(positions)], width=bar_width, label=f'Frequency of {categories[i]}', color = colors[i])

plt.xlabel('Frequency')
plt.ylabel('Count')
plt.title('Frequency of Mental Health Issues')
plt.xticks(positions + bar_width / 2, sorted(set(df['Anxiety'].tolist())))

legend = plt.legend()
for text, color in zip(legend.get_texts(), colors):
    text.set_color('#FFFFFF')

plt.show()

```

# ***Final Graphs / Bonus Tracks***

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data_path = 'mxmh_survey_results.csv'  
df = pd.read_csv(data_path)

mental_health_columns = ['Anxiety', 'Depression', 'Insomnia', 'OCD']
genre_columns = [col for col in df.columns if 'Frequency [' in col]
df_long = pd.melt(df, id_vars=mental_health_columns, value_vars=genre_columns, var_name='Genre', value_name='Frequency')
df_long['Genre'] = df_long['Genre'].str.extract(r'Frequency \[(.*?)\]')[0] 
frequency_to_ordinal = {
    'Never': 0,
    'Rarely': 1,
    'Sometimes': 2,
    'Frequently': 3,
    'Very frequently': 4
}
df_long['Frequency_Ordinal'] = df_long['Frequency'].map(frequency_to_ordinal)
df_grouped_anxiety = df_long.groupby(['Genre', 'Frequency_Ordinal'])['Anxiety'].mean().reset_index()

df_pivot_anxiety = df_grouped_anxiety.pivot(index="Frequency_Ordinal", columns="Genre", values="Anxiety") 

# heatmap!!!!
plt.figure(figsize=(20, 10))
heatmap_anxiety = sns.heatmap(df_pivot_anxiety, annot=True, cmap="gnuplot2", fmt=".2f")
heatmap_anxiety.set_title('Average Anxiety Score by Genre and Listening Frequency', size=20)
heatmap_anxiety.set_xlabel('Genre', size=15)
heatmap_anxiety.set_ylabel('Frequency (Ordinal)', size=15)
plt.xticks(rotation=45, ha="right", size=12)

plt.tight_layout()
plt.show()
```

### ***Bonus Track #3: OCD Orchestrative Behaviors***
### As shown here, we see Obsessive-Compulsive Disorder-based behavior peak in those who listen to primarily R&B music, and plummet in those who only listen to it sometimes-- an interesting conclusion that speaks volumes about R&B music-- but doesn't clarify it's effect nonetheless.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data_path = 'mxmh_survey_results.csv'  
df = pd.read_csv(data_path)

mental_health_columns = ['Anxiety', 'Depression', 'Insomnia', 'OCD']
genre_columns = [col for col in df.columns if 'Frequency [' in col]
df_long = pd.melt(df, id_vars=mental_health_columns, value_vars=genre_columns, var_name='Genre', value_name='Frequency')
df_long['Genre'] = df_long['Genre'].str.extract(r'Frequency \[(.*?)\]')[0] 
frequency_to_ordinal = {
    'Never': 0,
    'Rarely': 1,
    'Sometimes': 2,
    'Frequently': 3,
    'Very frequently': 4
}
df_long['Frequency_Ordinal'] = df_long['Frequency'].map(frequency_to_ordinal)
df_grouped_anxiety = df_long.groupby(['Genre', 'Frequency_Ordinal'])['Insomnia'].mean().reset_index()
df_pivot_anxiety = df_grouped_anxiety.pivot(index="Frequency_Ordinal", columns="Genre", values="Insomnia") 

# heatmap!!!!
plt.figure(figsize=(20, 10))
heatmap_anxiety = sns.heatmap(df_pivot_anxiety, annot=True, cmap="jet", fmt=".2f")
heatmap_anxiety.set_title('Average Insomnia Score by Genre and Listening Frequency', size=20)
heatmap_anxiety.set_xlabel('Genre', size=15)
heatmap_anxiety.set_ylabel('Frequency (Ordinal)', size=15)
plt.xticks(rotation=45, ha="right", size=12)

plt.tight_layout()
plt.show()
```

### ***Bonus Track #4: Insomnia Instrumentals***
### This final heatmap displays a pattern in those who favor Jazz to also display signs of Insomnia, not like those who listen to country music, who seem to not experience any sleep deprivation of the sort, an interesting conclusion considering the dissonance found in both music genres. 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data_path = 'mxmh_survey_results.csv'
df = pd.read_csv(data_path)

df_long = pd.melt(df, id_vars=["Music effects"], value_vars=[col for col in df.columns if col.startswith("Frequency")], 
                  var_name='Genre', value_name='Frequency')

music_effects_mapping = {'No effect': 0, 'Improve': .5, 'Worsen': 1}
df_long['Music effects'] = df_long['Music effects'].map(music_effects_mapping)

plt.figure(figsize=(12, 8))
sns.barplot(data=df_long, x='Music effects', y='Frequency', hue='Genre', ci=None)

plt.title('Averaged Out Reported Music effects by Genre Listening Frequency')
plt.xlabel('Music Effects')
plt.xlim(0,1)
plt.xticks([0, 0.5, 1], ['Improve', 'No effect', 'Worsen'])
plt.ylabel('Frequency')

plt.legend(fontsize='5')

plt.show()

```

### ***Bonus Track #5: Self-Reported Musical Impact***
### When all genres are taken into consideration and averaged out, it appears that the composition reveal that the general average regardless of genre and frequency of it being listened to, all falls between *No Effect* and *Improve*, with those who frequently listen to gospel music being the only group to record that music's general effect is between worsening their mental well-being and not having an effect on it.
